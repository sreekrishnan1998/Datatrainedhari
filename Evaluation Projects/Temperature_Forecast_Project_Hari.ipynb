```python
#treating outliers
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
IQR = q3 - q1
df1 = df[~((df < (q1 - 1.5 * IQR)) | (df > (q3 + 1.5 * IQR))).any(axis=1)]
df1.shape
```

#Predicting next day t_min

```python
#Selecting feature and target variable
x=df1.drop(['Next_Tmin'],axis=1)
y=df1['Next_Tmin']
```

```python
#Standadizing datavalues
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
#VIF Factor
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif['Features'] = x.columns
vif
```

```python
x=x.drop('LDAPS_Tmax_lapse',axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
#finding best random state
maxAcc=0
maxRs=0

for i in range(1,200):
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=i)
  lr=LinearRegression()
  lr.fit(x_train,y_train)
  pred=lr.predict(x_test)
  acc=r2_score(y_test,pred)
  if acc>maxAcc:
    maxAcc=acc
    maxRs=i
print('Best Accuracy', maxAcc, 'random_state is', maxRs)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the models
models = {
    "RandomForestRegressor": RandomForestRegressor(),
    "KNeighborsRegressor": KNeighborsRegressor(),
    "GradientBoostingRegressor": GradientBoostingRegressor(),
    "Lasso": Lasso(),
    "Ridge": Ridge(),
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "SVR": SVR(),
    "ExtraTreesRegressor": ExtraTreesRegressor(),
}

# Define the hyperparameter grids
param_grids = {
    "RandomForestRegressor": {'n_estimators': [100, 200, 300],'max_depth': [None, 5, 10],'min_samples_split': [2, 5, 10],'min_samples_leaf': [1, 2, 4]},
    "KNeighborsRegressor": {"n_neighbors": [5, 10, 15]},
    "GradientBoostingRegressor": {'n_estimators': [50, 100, 200],'learning_rate': [0.01, 0.1, 0.2],'max_depth': [3, 4, 5], 'min_samples_split': [2, 5, 10],'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 0.9, 1.0]},
    "Lasso": {"alpha": [0.1, 0.5, 1]},
    "Ridge": {"alpha": [0.1, 0.5, 1]},
    "DecisionTreeRegressor": {"max_depth": [None, 5, 10]},
    "SVR": {"kernel": ["linear", "rbf"], "C": [0.1, 1, 10]},
    "ExtraTreesRegressor": {'n_estimators': [100, 200, 300],'max_depth': [None, 5, 10],'min_samples_split': [2, 5, 10],'min_samples_leaf': [1, 2, 4]}
}

# Perform hyperparameter tuning for each model
for model_name, model in models.items():
    grid_search = GridSearchCV(model, param_grids[model_name], cv=5)
    grid_search.fit(x_train, y_train)

    # Print the best parameters and score
    print(f"Model: {model_name}")
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best estimator: {grid_search.best_estimator_}")
    print(f"Best score: {grid_search.best_score_}")
    print()
```
